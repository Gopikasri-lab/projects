# Train and Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,Y_train,Y_valid= train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score,f1_score
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Pred_cv=model.predict(X_valid)
```

```python
print("Accuracy Score = ",accuracy_score(Y_valid,Pred_cv))
```

```python
print("F1 score = ", f1_score(Y_valid,Pred_cv))
```

```python
pred_test=model.predict(test)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
i=0
mean=0
fmean=0
kf=StratifiedKFold(n_splits=5)
for train_i, test_i in kf.split(X,Y):
    print("\n {} of Kfold {}".format(i,kf.n_splits))
    xtr,xvl = X.loc[train_i],X.loc[test_i]
    ytr,yvl = Y[train_i],Y[test_i]
    model=LogisticRegression(random_state=1)
    model.fit(xtr,ytr)
    pred_test = model.predict(xvl)
    score=accuracy_score(pred_test,yvl)
    f1score=f1_score(pred_test,yvl)
    mean+=score
    fmean+=f1score
    print("\n accuracy_score",score)
    print("\n f1 score",f1score)
    i+=1
    pred_test_f=model.predict(test)
    pred=model.predict_proba(xvl)[:,1]

print("\n Final mean score")
print("\n Mean Validation Accuracy",mean/(i-1))
print("\n Mean Validation F1 score",fmean/(i-1))

```

```python
from sklearn import metrics
fpr,tpr,_=metrics.roc_curve(yvl,pred)
auc=metrics.roc_auc_score(yvl,pred)
plt.figure(figsize=(12,8))
plt.plot(fpr,tpr,label="validation, auc"+str(auc))
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend(loc=4)
plt.show()
```
